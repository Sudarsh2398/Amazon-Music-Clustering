```python
# import k means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []
for i in range(1,20):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(req_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,20), wcss)
plt.title('Elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=7, random_state=42)
kmeans.fit(req_df)
```

```python
df['Clusters'] = kmeans.predict(req_df)
df
```

```python
df['Clusters'].value_counts()
```

```python
# silhouette score
from sklearn.metrics import silhouette_score
score = silhouette_score(req_df, kmeans.labels_)
score
```

```python
from sklearn.metrics import davies_bouldin_score
db_score = davies_bouldin_score(req_df, kmeans.labels_)
db_score
```

```python
# cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Recomendation function

def Recomendation(name_song,top_n=5):
  if name_song not in df['name_song'].values:
    return 'Songs not found'

  idx=df[df['name_song']==name_song].index[0]
  cluster_id=df.loc[idx,'Clusters']

  # Find all movies in the same cluster
  clustered_song=df[df['Clusters']==cluster_id]

  # Compute the Cosine Similarity
  cosin_sim=cosine_similarity(req_df.iloc[idx].values.reshape(1,-1),req_df.loc[clustered_song.index]).flatten()
  # cosin_sim=cosine_similarity(tfidf_matrix[idx],tfidf_matrix[clustered_movie.index]).flatten()
  sim_score=list(zip(clustered_song['name_song'],cosin_sim))

  # Sor the Movie based on similarity score
  sim_score=sorted(sim_score,key=lambda x:x[1],reverse=True)
  recommendation=[m for m,s in sim_score[1:top_n+1]]
  return recommendation
```
